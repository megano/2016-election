```python
import pandas as pd
import numpy as np
import datetime

xls_file = pd.read_excel('../data/16_gen_ab_ev_list_10_23_16.xlsx')
df = pd.DataFrame(xls_file, columns = ['voter_id', 'name_last', 'Category', 'date_entered', 'Original Party'])
df.head()
```

```python
# Select Category column
df['Category']
```

```python
# Select rows where Category column is early voter (EV)
df = df[df['Category'] == "EV"]
```

```python
# Convert the date_entered values to datetime format
df['date_entered'] = pd.to_datetime(df['date_entered'], errors='ignore')
df.head()
```

```python
# rename date_entered column to datetime
df = df.rename(columns = {'date_entered':'datetime'})
df.head()
```

```python
# aggregate date_entered by hour and minute
times = pd.DatetimeIndex(df.datetime)
grouped = df.groupby([times.hour, times.minute])
grouped.head()
```
